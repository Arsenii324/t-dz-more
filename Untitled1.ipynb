```python
import torch
import transformer_lens.utils as utils
from transformer_lens import HookedTransformer
```

```python
try:
    model = HookedTransformer.from_pretrained("EleutherAI/pythia-410m")
except Exception as e:
    print(e)
```

```python
model
```

```python
add_1 = 18
add_2 = 28
add_1_s = str(add_1)
add_2_s = str(add_2)

sum_clean = add_1 + add_2
sum_clean_s = str(sum_clean)

clean_prompt = ' '.join(add_1_s) + ' + ' + ' '.join(add_2_s) + ' = ' + ' '.join(sum_clean_s)

add_1_corrupted = 11
add_2_corrupted = 21
add_1_corrupted_s = str(add_1_corrupted)
add_2_corrupted_s = str(add_2_corrupted)

sum_corrupted = add_1_corrupted + add_2_corrupted
sum_corrupted_s = str(sum_corrupted)

corrupted_prompt = ' '.join(add_1_corrupted_s) + ' + ' + ' '.join(add_2_corrupted_s) + ' = ' + ' '.join(sum_corrupted_s)


clean_tokens = model.to_tokens(clean_prompt)
corrupted_tokens = model.to_tokens(corrupted_prompt)


clean_logits, clean_cache = model.run_with_cache(clean_tokens)
corrupted_logits, corrupted_cache = model.run_with_cache(corrupted_tokens)
```

```python
def patch_head_activation_hook(activation, hook, head_index):

    activation[:, :, head_index, :] = corrupted_cache[hook.name][:, :, head_index, :]
    return activation
```

```python
layer_to_patch = 5
head_to_patch = 2
hook_point = f"blocks.{layer_to_patch}.attn.hook_z"

patched_logits = model.run_with_hooks(
    clean_tokens,
    fwd_hooks=[
        (hook_point, lambda act, hook: patch_head_activation_hook(act, hook, head_to_patch))
    ]
)
```

```python
import torch
import torch.nn.functional as F

def calculate_pdf_indirect_effect(clean_logits, patched_logits, r_token, r_prime_token):
    clean_last_logits = clean_logits[0, -1, :]
    patched_last_logits = patched_logits[0, -1, :]

    p = F.softmax(clean_last_logits, dim=-1)
    p_star = F.softmax(patched_last_logits, dim=-1)

    term_1 = (p_star[r_prime_token] - p[r_prime_token]) / p[r_prime_token]

    term_2 = (p[r_token] - p_star[r_token]) / p_star[r_token]

    return ((term_1 + term_2) / 2).item()

r_token = model.to_single_token("4")
r_prime_token = model.to_single_token("3")

ie_sanity = calculate_pdf_indirect_effect(
    clean_logits,
    corrupted_logits,
    r_token,
    r_prime_token
)

ie_value = calculate_pdf_indirect_effect(
    clean_logits,
    patched_logits,
    r_token,
    r_prime_token
)

print(f"Sanity Check, Prob Diff Clean vs Corrupt: {ie_sanity:.4f}")
print(f"Indirect Effect: {ie_value:.4f}")
```

```python
for layer_to_patch in range(3, 7):
    for head_to_patch in (1, 2):
        hook_point = f"blocks.{layer_to_patch}.attn.hook_z"

        patched_logits = model.run_with_hooks(
            clean_tokens,
            fwd_hooks=[
                (hook_point, lambda act, hook: patch_head_activation_hook(act, hook, head_to_patch))
            ]
        )

        ie_value = calculate_pdf_indirect_effect(
            clean_logits,
            patched_logits,
            r_token,
            r_prime_token
        )

        # print(f"Sanity Check, Prob Diff Clean vs Corrupt: {ie_sanity:.4f}")
        print(f"Indirect Effect: {ie_value:.4f}")
```
